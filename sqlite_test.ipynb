```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy import create_engine, text, inspect, func
```

```python
# Create engine using the `demographics.sqlite` database file
engine = create_engine("sqlite:///ofsted_results.sqlite")
```

```python
inspector = inspect(engine)
```

```python
table_names = inspector.get_table_names()
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
# Base.prepare(autoload_with=engine, reflect=True)
```

```python
# Using alternative method
# Manually create classes for each table using the original table names
for table_name in table_names:
    table = sqlalchemy.Table(table_name, Base.metadata, autoload_with=engine)
    mapped_class = type(table_name, (Base,), {'__table__': table})
    setattr(Base.classes, table_name, mapped_class)
```

```python
# Print all of the classes mapped to the Base
Base.classes.keys()
```
