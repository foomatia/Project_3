```python
import pandas as pd

# Read the data into a Pandas DataFrame with explicit encoding
# Assuming 'utf-8' encoding, change it if needed
df = pd.read_csv("raw_dataset/State_funded_schools_inspections_and_outcomes_as_at_31_August_2023.csv", encoding='latin-1')

# Display the first few rows of the DataFrame
df.head()
```

```python
# Reduced df of columns
reduced_df = df[["Web Link","URN","School name","Ofsted phase","Ofsted region","Local authority",
                 "Postcode","Total number of pupils","Inspection start date","Inspection end date",
                 "Publication date","Does the latest graded inspection relate to the URN of the current school?",
                 "Overall effectiveness","Category of concern","Quality of education","Behaviour and attitudes",
                 "Personal development","Effectiveness of leadership and management","Safeguarding is effective?"]]

reduced_df.head()
```

```python
# Show dtypes
reduced_df.info()
```

### Grades
1 = Outstanding <br>
2 = Good <br>
3 = Requires Improvement <br>
4 = Inadequate <br>
9/NaN = Insufficient Evidence <br>

```python
# Mass value change of Ofsted Grades for ease of understanding
reduced_df[['Overall effectiveness','Quality of education',
            'Behaviour and attitudes','Personal development',
            'Effectiveness of leadership and management']] = reduced_df[['Overall effectiveness','Quality of education',
                                                                         'Behaviour and attitudes','Personal development',
                                                                         'Effectiveness of leadership and management']].replace({1.0:"Outstanding",
                                                                                                                                2.0:"Good",
                                                                                                                                3.0:"Requires Improvement",
                                                                                                                                4.0:"Inadequate",
                                                                                                                                9.0:"Insufficient Evidence"})

reduced_df.head()
```

```python
# Alter 'Safeguarding is effective?' column values
reduced_df['Safeguarding is effective?'] = reduced_df['Safeguarding is effective?'].replace({"9":"Insufficient Evidence"})

reduced_df.head()
```

```python
# Filter the data to only Primary or Seconday Phases
filtered_df = reduced_df.loc[(reduced_df["Ofsted phase"] == "Primary") | (reduced_df["Ofsted phase"] == "Secondary")]

# Filter the data to only West Midlands Region
filtered_df = filtered_df.loc[reduced_df["Ofsted region"] == "West Midlands"]

filtered_df.head()
```

```python
# Change NULL values in Num of Pupils to 0
filtered_df['Total number of pupils'] = filtered_df['Total number of pupils'].fillna(0)

# Alter column data types
filtered_df = filtered_df.astype({"URN":"int64",
                               "Total number of pupils":"int64",
                                 "Inspection start date":"datetime64",
                                 "Inspection end date":"datetime64",
                                 "Publication date":"datetime64"})

filtered_df.dtypes
```

```python
filtered_df = filtered_df.reset_index(drop=True)

filtered_df.head()
```

```python
# Create df containing only school name, URN and postcode
school_postal = filtered_df[["URN","School name","Postcode"]]

school_postal.to_csv("Resources/school_postal.csv", index=False)
```

```python
from geopy.geocoders import Nominatim
import pandas as pd

# Assuming df is your DataFrame
df = pd.read_csv("Resources/State_funded_schools_inspections_and_outcomes_as_at_31_August_2023.csv", encoding='latin-1')

# Create a geolocator using Nominatim
geolocator = Nominatim(user_agent="school_geocoder")

# Function to get latitude and longitude from address
def get_lat_lon(Postcode):
    location = geolocator.geocode(Postcode)
    return (location.latitude, location.longitude) if location else (None, None)

# Apply the function to the 'Postcode' column
df['Latitude'], df['Longitude'] = zip(*df['Postcode'].apply(get_lat_lon))

# Display the DataFrame with School name, Latitude, and Longitude columns
result_df = df[['School name', 'Latitude', 'Longitude']]
print(result_df)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim

# Assuming df is your DataFrame
df = pd.read_csv("Resources/State_funded_schools_inspections_and_outcomes_as_at_31_August_2023.csv", encoding='ISO-8859-1')

# Define the target region and local authority
target_region = 'West Midlands'
target_local_authority = 'Birmingham'

# Filter the DataFrame based on both conditions
filtered1_df = df[(df['Ofsted region'] == target_region) & (df['Local authority'] == target_local_authority)]

# Create a geolocator using Nominatim
geolocator = Nominatim(user_agent="school_geocoder")

# Function to get latitude and longitude from address (Postcode)
def get_lat_lon(postcode):
    location = geolocator.geocode(postcode)
    return (location.latitude, location.longitude) if location else (None, None)

# Apply the function to the 'Postcode' column
filtered1_df['Latitude'], filtered1_df['Longitude'] = zip(*filtered1_df['Postcode'].apply(get_lat_lon))

# Display the first few rows of the updated DataFrame
print(filtered1_df.head())
```
